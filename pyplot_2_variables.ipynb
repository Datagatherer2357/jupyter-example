### Two variable plots

Adapted from: https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html

```python
# Relationships between x and y

import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4]) # y values (Always assumes y axis first)
plt.ylabel('y axis')
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # x vs y arrays to plot

# pyplot by default interpolates (joins via lines) the points
```

```python
# Format style:

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # r is red o is the marker
plt.axis([0, 6, 0, 20])
```

$$ y = 2x $$

```python
x = np.linspace(0.0, 100.0, 75) # 0 - 100 with 1000 points

y1 = 2.0 *x + 5.0 # last value is the intercept
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0 # slope is negative here

plt.plot(x, y1, 'b.')
plt.plot(x, y2, 'r.')
plt.plot(x, y3, 'g.') #le8 is 10 to the power of 8
```
